## Linear equations

In your example equation:

$$y' + 2xy = 10x$$

We go about solving by taking what's called an "ansatz" (it means "guess").  Because we looked it up, we know the right guess is

$$y = e^u$$

for some $u$ that we need to figure out.  Let's put that in to our equation, noting that the chain rule says

$$y' = C e^u \frac{du}{dx}$$

We're going to get the general solution first (the one where the right hand side is set to zero)

$$\frac{du}{dx} e^u + 2x\,e^u = 0$$

Well, the only way that's going to work is if the things multiplying the $e^u$ cancel, so we need

$$\frac{du}{dx} = - 2x$$

That tells us that the $u$ we need is (by integrating)

$$u = -x^2 + C$$

Giving a solution term of

$$y = C e^{-x^2}$$

Ok, so far so good, that's the general solution.  Now for the specific - we need something that, when taking derivatives as listed, gives us $10x$.  That looks like a job for some sort of polynomial.  Let's try

$$y = a x^2 + b x + c$$

which would have

$$y' = 2a x + b$$

Plugging _that_ in, we get

$$(2ax + b) + 2x*(a x^2 + b x + c) = 10 x$$

So

$$2a x^3 + 2b x^2 + (2a + 2c) x + b = 10 x$$

Ok, matching up coefficiencts, we get

$$2a = 0 \qquad 2b = 0 \qquad (2a + 2c) = 10 \qquad b = 0$$

So it looks like we didn't need most of that, since $a=0$ and $b=0$.  That leaves us with $2c = 10$, so $c=5$

This leaves a solution of

$$y = 5$$

Now, we add the general and the specific to get

$$y = C e^{-x^2} + 5$$

Now, let's check that it works

$$y' = (-2x) C e^{-x^2}$$

Let's plug these in to our original equation

$$y' + 2xy = 10x$$

$$(-2x) C e^{-x^2} + 2x (C e^{-x^2} + 5) = 10 x$$

Let's subtract the 10 x from both sides, and multiply out the second term

$$(-2x) C e^{-x^2} + 2x C e^{-x^2} + 10 x - 10 x = 0$$

$$0 = 0$$

So everything worked out
